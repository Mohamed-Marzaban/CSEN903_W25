# Data Engineering Questions


# Quesiton 1
#Which city is best for each traveler type? For each traveler type, recommend the best city based on the given reviews.

```python
merged_df = pd.merge(reviews_df, users_df, on='user_id')
merged_df = pd.merge(merged_df, hotels_df, on='hotel_id')
merged_df.head()
```

```python
pivot_df = merged_df.groupby(['traveller_type', 'city'])['score_overall'].mean().unstack()

plt.figure(figsize=(14, 6))
sns.heatmap(
    pivot_df,
    annot=True,
    fmt=".2f",
    cmap='YlGnBu',
    linewidths=0.5,
    cbar_kws={'label': 'Average Overall-Review Score'}
)

# Titles and labels
plt.title('Average Overall-Review Score by Traveler Type and City', fontsize=16, fontweight='bold')
plt.xlabel('City', fontsize=12)
plt.ylabel('Traveler Type', fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Based on the heatmap provided:
# 1-Amsterdam is the best city for solo travelers with average overall review score of 9.11
# 2-Dubai is the best city for famiy travelers with average overall review score of 9.21
# 3-Amsterdam is the best city for couple travelers with average overall review score of 9.10
# 4-Dubai is the best city for business travelers with average overall review score of 8.97

# Question 2
# What are the top 3 countries with the best value-for-money score per traveler’s age group?

```python
# Step 1: Group and pivot the data
pivot_df_1 = merged_df.groupby(['age_group', 'country_y'])['score_value_for_money'].mean().unstack()


plt.figure(figsize=(14, 6))


sns.heatmap(
    pivot_df_1,
    annot=True,
    fmt=".2f",
    cmap='YlOrBr',
    linewidths=0.5,
    cbar_kws={'label': 'Average Score for Value for Money'}
)


plt.title('Average "Value for Money" Score by Age Group and Country', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Age Group', fontsize=12)


plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Top 3 countires for Each Age Group

```python
top_3_18_24 = pd.Series(pivot_df_1.iloc[0].sort_values(ascending=False).head(3))
top_3_25_34 = pd.Series(pivot_df_1.iloc[1].sort_values(ascending=False).head(3))
top_3_35_44 = pd.Series(pivot_df_1.iloc[2].sort_values(ascending=False).head(3))
top_3_45_54 = pd.Series(pivot_df_1.iloc[3].sort_values(ascending=False).head(3))
top_3_55 = pd.Series(pivot_df_1.iloc[4].sort_values(ascending=False).head(3))

```

```python
# Your data dictionary (same as before)
top_3_dict = {
    '18–24': top_3_18_24,
    '25–34': top_3_25_34,
    '35–44': top_3_35_44,
    '45–54': top_3_45_54,
    '55+': top_3_55
}

# Step 1: Setup
age_groups = list(top_3_dict.keys())
x = np.arange(len(age_groups))
width = 0.25

# Step 2: Extract bars and labels
bars = []
labels = []
for i in range(3):
    scores = [list(top_3_dict[age].values)[i] for age in age_groups]
    countries = [list(top_3_dict[age].keys())[i] for age in age_groups]
    bars.append(scores)
    labels.append(countries)

# Step 3: Define nice custom colors
colors = ['#1f77b4', '#2ca02c', '#ff7f0e']  # blue, green, orange

# Step 4: Plot grouped bars with colors
plt.figure(figsize=(12, 6))
bar1 = plt.bar(x - width, bars[0], width, color=colors[0], label='Top 1')
bar2 = plt.bar(x,         bars[1], width, color=colors[1], label='Top 2')
bar3 = plt.bar(x + width, bars[2], width, color=colors[2], label='Top 3')

# Step 5: Add labels on top of each bar
for i in range(len(age_groups)):
    plt.text(x[i] - width, bars[0][i] + 0.05, labels[0][i], ha='center', fontsize=9, rotation=45)
    plt.text(x[i],         bars[1][i] + 0.05, labels[1][i], ha='center', fontsize=9, rotation=45)
    plt.text(x[i] + width, bars[2][i] + 0.05, labels[2][i], ha='center', fontsize=9, rotation=45)

# Step 6: Style and display
plt.xticks(x, age_groups)
plt.ylabel('Average Score (Value for Money)', fontsize=11)
plt.xlabel('Age Group', fontsize=11)
plt.title('Top 3 Countries per Age Group by Value for Money', fontsize=14, fontweight='bold')
plt.legend(title='Ranking', loc='upper left')
plt.ylim(0, 10)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
